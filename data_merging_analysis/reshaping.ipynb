```python
def generate_sample_id(row):
    if pd.notnull(row['sampling_location']):
        # If sampling_location exists, combine it with the sampling_date
        return f"{row['sampling_date']}_{row['sampling_location']}"
    elif pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        # If latitude and longitude exist, combine them with the sampling_date
        return f"{row['sampling_date']}_{row['latitude']}_{row['longitude']}"
    elif pd.notnull(row['pws_id']):
        # If sampling_location, latitude, and longitude are null, use pwd_id
        return f"{row['sampling_date']}_{row['pws_id']}"
    else:
        # If none of the above exist, return NaN (no way to generate ID)
        return np.nan
```

```python
# Fill only rows where 'sample_id' is currently null
us_pfas_dw.loc[us_pfas_dw['sample_id'].isnull(), 'sample_id'] = (
    us_pfas_dw[us_pfas_dw['sample_id'].isnull()]
    .apply(generate_sample_id, axis=1)
)
```
