```python
def extract_acronym(compound):
    match = re.search(r'\((\*?[A-Z0-9\-]+)\)', compound)
    if match:
        return match.group(1).replace('*', '')
    else:
        acronym_match = re.search(r'^[A-Z0-9\-]+$', compound.strip())
        if acronym_match:
            return acronym_match.group(0)
        else:
            # Return None if no acronym is found
            return np.nan
```

```python
vt_df['analyte'] = vt_df['Compound'].apply(extract_acronym)
```

```python
vt_df['analyte'] = vt_df['analyte'].fillna('PFHPA+PFHXS+PFNA+PFOS+PFOA')
```

```python
vt_df
```

```python
def process_result(value):
    if value == 'ND':
        return float(0)
    else:
        return float(value)
```

```python
vt_df['concentration'] = vt_df['Concentration (ppt)'].apply(process_result)
```
