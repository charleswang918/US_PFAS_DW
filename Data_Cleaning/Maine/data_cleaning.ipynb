```python
me_df4 = pd.read_csv('tabula-PFASallResults.csv')
me_df4
```

```python
me_df4 = me_df4.drop([0, 1, 2])
```

```python
me_df4.columns = list(me_df4.iloc[0])
```

```python
me_df4 = me_df4.rename(columns={'2807)': 'PFHPA+PFHXS+PFNA+PFOS+PFOA+PFDA'})
```

```python
me_df4 = me_df4.drop([3])
```

```python
me_df4
```

```python
me_df4 = me_df4[me_df4['PWSID'] != 'PWSID']
```

```python
me_df4 = me_df4.replace('ND', 0)
```

```python
me_df4.columns = me_df4.columns.str.replace(r'\(|\)', '', regex=True)
```

```python
me_df4.index = range(0, len(me_df4))
```

```python
columns_to_repeat = ['PWSID', 'System Name', 'Facility', 'Date']
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(me_df4)):
    row_repeated = pd.concat([me_df4.iloc[[i]][columns_to_repeat]] * 14, ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)

df_repeated
```

```python
df_repeated['analyte'] = list(me_df4.columns)[4:18] * len(me_df4)
```

```python
df_repeated
```

```python
analyte_list = list(me_df4.columns)[4:18]
```

```python
me_df4[analyte_list]
```

```python
me_df4[analyte_list] = me_df4[analyte_list].astype(float)
```

```python
me_df4[analyte_list].values.flatten().tolist()
```

```python
len(me_df4[analyte_list].values.flatten().tolist())
```

```python
df_repeated['concentration'] = me_df4[analyte_list].values.flatten().tolist()
```

```python
def set_is_detected(concentration):
    if pd.isna(concentration):
        return np.nan
    return concentration != 0
```

```python
df_repeated['is_detected'] = df_repeated['concentration'].apply(set_is_detected)
```

```python
df_repeated
```
