```python
import pandas as pd
import os
import numpy as np
```

```python
tn_df1 = pd.read_excel('Export_Arnold AFB.xlsx')
```

```python
tn_df1
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
```

```python
df1['analyte'] = tn_df1['Analyte Abbreviation'].str.upper()
df1['concentration'] = tn_df1['Final Result'].fillna(0)
df1['sampling_location'] = tn_df1['Installation Name']
df1['DL'] = tn_df1['Limit of Detection']
df1['sampling_date'] = pd.to_datetime(tn_df1['Sample Date'])
df1['sample_id'] = tn_df1['Laboratory Sample ID']
df1['units'] = tn_df1['Final Units']
df1['state'] = tn_df1['State']
df1['analytical_methods'] = tn_df1['Analysis Method']
df1['data_source'] = 'DoD'
df1['source_type'] = 'drinking water'
```

```python
df1['is_detected'] = df1['concentration'] != 0
```

```python
df1
```

```python
tables = pd.read_excel('opp_pfas_samplingresults.xlsx', sheet_name = 'Sample Results')
```

```python
tables
```

```python
tables.columns
```

```python
new_columns = []
for col in tables.columns:
    if 'ng/L' in col:
        new_columns.append(col.split('(')[0].strip())
    else:
        new_columns.append(col)
new_columns
```

```python
tables.columns = new_columns
```

```python
tables.dtypes
```

```python
tables
```

```python
columns_to_fill = [
    'City', 'Public Utility System Name', 'Public Utility System ID', 
    'Public Utility System Population Served', 'Number of Sources Sampled'
]
```

```python
tables[columns_to_fill] = tables[columns_to_fill].fillna(method='ffill')
```

```python
tables
```

```python
analyte_list = list(tables.columns[10:])
analyte_list
```

```python
list(tables.columns[0:10])
```

```python
columns_to_repeat = list(tables.columns[0:10])
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(tables)):
    row_repeated = pd.concat([tables.iloc[[i]][columns_to_repeat]] * len(analyte_list), ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)

df_repeated
```

```python
df_repeated['analyte'] = analyte_list * len(tables)
```

```python
df_repeated
```

```python
tables = tables.replace({'Non-Detect': 0, 'Detected Below MRL': 0, 'Q': np.nan})
```

```python
tables[analyte_list] = tables[analyte_list].astype(float)
```

```python
tables.dtypes
```

```python
df_repeated['concentration'] = tables[analyte_list].values.flatten()
```

```python
df_repeated
```

```python
df2 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df2['city'] = df_repeated['City']
df2['water_system_name'] = df_repeated['Public Utility System Name']
df2['pws_id'] = df_repeated['Public Utility System ID']
df2['population'] = df_repeated['Public Utility System Population Served'].astype(int)
df2['source_type'] = df_repeated['Source Type']
df2['units'] = 'ng/L'
df2['note'] = df_repeated['Raw or Treated Water']
df2['sampling_date'] = pd.to_datetime(df_repeated['Sample Date'])
df2['analyte'] = df_repeated['analyte']
df2['concentration'] = df_repeated['concentration']
df2['state'] = 'Tennessee'
df2['data_source'] = 'TDEC'
```

```python
def set_is_detected(concentration):
    if pd.isna(concentration):
        return np.nan
    return concentration != 0
```

```python
df2['is_detected'] = df2['concentration'].apply(set_is_detected)
```

```python
df2
```

```python
df = pd.concat([df1, df2])
```

```python
df.index = range(0, len(df))
```

```python
df
```

```python
df.to_csv('tennessee_dw.csv')
```
