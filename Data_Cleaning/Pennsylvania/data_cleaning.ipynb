```python
tables = pd.read_csv('tabula-PFAS_Sampling_Final_Results_May_2021.csv')
```

```python
tables
```

```python
tables = tables.rename(columns={'BOL Sample # /\rELLE Job #' : 'BOL Sample#/ELLE Job#', 'Sum_2\r(PFOS +\rPFOA)': 'PFOS+PFOA'})
```

```python
list(tables.columns[7:26])
```

```python
columns_to_repeat = ['PWSID', 'PWS Name', 'County', 'Date\rCollected', 'Category']
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(tables)):
    row_repeated = pd.concat([tables.iloc[[i]][columns_to_repeat]] * len(list(tables.columns[7:26])), ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)

df_repeated
```

```python
df_repeated['analyte'] = list(tables.columns[7:26]) * len(tables)
```

```python
df_repeated['concentration'] = tables[list(tables.columns[7:26])].values.flatten().tolist()
```

```python
df_repeated['analyte'] = df_repeated['analyte'].str.upper()
```

```python
df_repeated = df_repeated.replace({'ND': float(0)})
```

```python
df_repeated.dtypes
```

```python
df_repeated['PWSID'] = df_repeated['PWSID'].astype(str)
```

```python
df_repeated['concentration'] = df_repeated['concentration'].astype(float)
```

```python
df_repeated['is_detected'] = df_repeated['concentration'] != 0
```

```python
df_repeated
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str 
}
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df1['analyte'] = df_repeated['analyte']
df1['concentration'] = df_repeated['concentration']
df1['pws_id'] = df_repeated['PWSID']
df1['water_system_name'] = df_repeated['PWS Name']
df1['sampling_date'] = pd.to_datetime(df_repeated['Date\rCollected'])
df1['units'] = 'ng/L'
df1['county'] = df_repeated['County']
df1['state'] = 'Pennsylvania'
df1['analytical_methods'] = 'EPA Method 537.1'
df1['data_source'] = 'DEP'
df1['source_type'] = 'drinking water'
df1['well_type'] = df_repeated['Category']
df1['note'] = 'TW/TI = Target Well (GW)/Target Intake (SW); BW/BI = Baseline Well (GW)/Baseline Intake (SW)'
```

```python
df1['is_detected'] = df1['concentration'] != 0
```
