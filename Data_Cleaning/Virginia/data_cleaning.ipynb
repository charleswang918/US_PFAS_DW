```python
analyte_list = ['PFBA', 'PFPEA', 'PFHXA', 'PFHPA', 'PFOA', 'PFNA', 'PFDA', 'PFDOA', 'PFBS', 'PFPES', 'PFHXS', 'PFHPS', 'PFOS', 
                '4:2FTS', '6:2FTS', '8:2FTS', 'HFPO-DA', 'ADONA', '9Cl-PF3ONS', '11Cl-PF3OUDS', 'NFDHA', 'PFEESA', 'PFMBA', 'PFMPA', 'PFUNA']
```

```python
len(analyte_list)
```

```python
columns_to_repeat = ['Area', 'House Number Street Name', 'Sample Collected', 'Analysis Date', 'MMA Received Results', 'Well Class']
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(tables)):
    row_repeated = pd.concat([tables.iloc[[i]][columns_to_repeat]] * len(analyte_list), ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)

df_repeated
```

```python
df_repeated['analyte'] = analyte_list * len(tables)
```

```python
df_repeated['concentration'] = tables['Raw Result'].values.flatten().tolist()
```

```python
df_repeated['DL'] = tables['LOD'].values.flatten().tolist()
```

```python
df_repeated
```

```python
filtered_df = df_repeated[df_repeated['Sample Collected'].str.contains('&', na=False)]
filtered_df
```

```python
def extract_correct_date(row):
    if '&' in row['Sample Collected']:
        date_part1, date_part2 = row['Sample Collected'].split('&')
        date1 = date_part1.strip()
        date2 = date_part2.split('(')[0].strip()
        if '6:2 FTS' in row['analyte']:
            return date2
        else:
            return date1
    else:
        return row['Sample Collected']
```

```python
df_repeated.loc[filtered_df.index, 'Sample Collected'] = filtered_df.apply(extract_correct_date, axis=1)
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df1['sampling_location'] = df_repeated['House Number Street Name']
df1['analyte'] = df_repeated['analyte']
df1['concentration'] = df_repeated['concentration']
df1['DL'] = df_repeated['DL']
df1['sampling_date'] = pd.to_datetime(df_repeated['Sample Collected'])
df1['units'] = 'ng/L'
df1['county'] = 'Henrico'
df1['state'] = 'Virginia'
df1['analytical_methods'] = 'USEPA Method 533 DW'
df1['well_type'] = df_repeated['Well Class']
df1['data_source'] = 'VIRGINIA DEPARTMENT OF HEALTH'
df1['source_type'] = 'drinking water'
```
