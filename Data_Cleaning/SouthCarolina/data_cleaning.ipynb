```python
import pandas as pd
import os
import numpy as np
import camelot
```

```python
tables = camelot.read_pdf('T13 Other PFAS by 537.1.pdf', pages = 'all')
```

```python
for i in range(len(tables)):
    print(len(tables[i].df.columns))
```

```python
a1 = tables[0].df
```

```python
a1 = a1.drop([0, 1, 3])
```

```python
a1
```

```python
col_list = list(a1.iloc[0].str.upper())
```

```python
col_list[0] = 'pws_id'
col_list[1] = 'water system name'
col_list[2] = 'sampling_date'
```

```python
a1.columns = col_list
```

```python
a1 = a1.drop([2])
```

```python
a2 = tables[10].df
```

```python
a2 = a2.drop([53, 54, 55, 56])
```

```python
a2.columns = col_list
```

```python
sc_df1 = pd.DataFrame(columns = a1.columns)

for i in range(1, len(tables) - 1):
    df = tables[i].df
    df.columns = a1.columns
    sc_df1 = pd.concat([sc_df1, df])
sc_df1
```

```python
sc_df1 = pd.concat([a1, sc_df1, a2])
```

```python
sc_df1
```

```python
sc_df1 = sc_df1.replace(['*', '* c.'], float(0))
```

```python
sc_df1['PFBS'] = sc_df1['PFBS'].astype(float)
sc_df1['PFHXA'] = sc_df1['PFHXA'].astype(float)
sc_df1['PFHXS'] = sc_df1['PFHXS'].astype(float)
sc_df1['PFHPA'] = sc_df1['PFHPA'].astype(float)
```

```python
sc_df1.dtypes
```

```python
sc_df1.index = range(0, len(sc_df1))
```

```python
sc_df1
```

```python
analyte_list = list(sc_df1.columns[3:])
```

```python
columns_to_repeat = ['pws_id', 'water system name', 'sampling_date']
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(sc_df1)):
    row_repeated = pd.concat([sc_df1.iloc[[i]][columns_to_repeat]] * len(analyte_list), ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)
```

```python
df_repeated
```

```python
df_repeated['analyte'] = analyte_list * len(sc_df1)
df_repeated['concentration'] = sc_df1[analyte_list].values.flatten().tolist()
```

```python
df_repeated['is_detected'] = df_repeated['concentration'] != 0
```

```python
df_repeated
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str 
}
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df1['analyte'] = df_repeated['analyte']
df1['concentration'] = df_repeated['concentration']
df1['pws_id'] = df_repeated['pws_id']
df1['water_system_name'] = df_repeated['water system name']
df1['sampling_date'] = pd.to_datetime(df_repeated['sampling_date'])
df1['units'] = 'ng/L'
df1['state'] = 'South Carolina'
df1['analytical_methods'] = 'EPA Method 537.1'
df1['data_source'] = 'DHEC'
df1['source_type'] = 'drinking water'
df1['is_detected'] = df_repeated['is_detected']
```
