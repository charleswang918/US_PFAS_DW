```python
col_df = pd.DataFrame()
for i in range(0, len(tables_2020), 2):
    tem_df = pd.DataFrame()
    df = tables_2020[i].df
    df = df.drop(df.index[-1])
    df = df.replace('', np.nan).dropna(axis=1, how='all')
    tem_df['location'] = [df.iloc[1, 0]]
    tem_df['sample_id'] = [df.iloc[3, 0]]
    tem_df['sample_date'] = [df.iloc[1, 1]]
    col_df = pd.concat([col_df, tem_df])
col_df
```

```python
col_df['sample_id'] = col_df['sample_id'].str.split(':').str[1].str.strip()
col_df['location'] = col_df['location'].str.split(':').str[1].str.strip()
col_df['sample_date'] = col_df['sample_date'].apply(lambda x: x.split(': ', 1)[1])
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
df = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df['sample_id'] = nd_df['sample_id']
df['sampling_location'] = nd_df['location']
df['data_source'] = 'DEQ'
df['sampling_date'] = pd.to_datetime(nd_df['sample_date'])
df['units'] = nd_df['Units']
df['source_type'] = 'drinking water'
df['state'] = 'North Dakota'
df['DL'] = nd_df['DL']
df['analytical_methods'] = 'EPA 537.1'
df['analyte'] = nd_df['Parameter'].str.upper()
df['concentration'] = nd_df['Result']
```

```python
def set_is_detected(concentration):
    if pd.isna(concentration):
        return np.nan
    return concentration != 0
```

```python
df['is_detected'] = df['concentration'].apply(set_is_detected)
```
