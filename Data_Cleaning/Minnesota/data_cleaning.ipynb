```python
import pandas as pd
import os
import numpy as np
```

```python
os.getcwd()
```

```python
mn_df1 = pd.read_excel('Export_Camp Ripley.xlsx')
```

```python
mn_df1
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
```

```python
df1['analyte'] = mn_df1['Analyte Abbreviation'].str.upper()
df1['concentration'] = mn_df1['Final Result'].fillna(0)
df1['sampling_location'] = mn_df1['Installation Name']
df1['DL'] = mn_df1['Limit of Detection']
df1['sampling_date'] = pd.to_datetime(mn_df1['Sample Date'])
df1['sample_id'] = mn_df1['Laboratory Sample ID']
df1['units'] = mn_df1['Final Units']
df1['state'] = mn_df1['State']
df1['analytical_methods'] = mn_df1['Analysis Method']
df1['data_source'] = 'DoD'
df1['source_type'] = 'drinking water'
```

```python
df1['is_detected'] = df1['concentration'] != 0
```

```python
df1
```

```python
mn_df2 = pd.read_excel('Export_Camp Ripley (Installation Wide PA) (Western AOI SI).xlsx')
```

```python
mn_df2
```

```python
df2 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df2['analyte'] = mn_df2['Analyte Abbreviation'].str.upper()
df2['concentration'] = mn_df2['Final Result'].fillna(0)
df2['sampling_location'] = mn_df2['Installation Name']
df2['DL'] = mn_df2['Limit of Detection']
df2['sampling_date'] = pd.to_datetime(mn_df2['Sample Date'])
df2['sample_id'] = mn_df2['Laboratory Sample ID']
df2['units'] = mn_df2['Final Units']
df2['state'] = mn_df2['State']
df2['analytical_methods'] = mn_df2['Analysis Method']
df2['data_source'] = 'DoD'
df2['source_type'] = 'drinking water'
```

```python
df = pd.concat([df1, df2])
```

```python
df.to_csv('minnesota_dw.csv')
```
