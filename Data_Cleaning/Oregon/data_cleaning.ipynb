```python
import pandas as pd
import os
import numpy as np
```

```python
os.getcwd()
```

```python
or_df1 = pd.read_csv('PFAS Data.csv')
```

```python
or_df1
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
```

```python
def process_result(value):
    if value == 'ND':
        return 0
    else:
        return float(value)
```

```python
or_df1['result'] = or_df1['Result (ng/L)'].apply(process_result)
```

```python
or_df1
```

```python
df1['sample_id'] = or_df1['Sample ID']
df1['data_source'] = 'OHA'
df1['sampling_date'] = pd.to_datetime(or_df1['Sample Date'])
df1['pws_id'] = or_df1['PWS']
df1['water_system_name'] = or_df1['PWS Name']
df1['units'] = 'ng/L'
df1['county'] = or_df1['County Served']
df1['MRL'] = or_df1['MRL (ng/L)']
df1['analyte'] = or_df1['Analyte Abbreviation'].str.upper()
df1['concentration'] = or_df1['result']
df1['source_type'] = 'drinking water'
df1['state'] = 'Oregon'
```

```python
df1['is_detected'] = or_df1['result'] != 0
```

```python
df1
```

```python
or_df2 = pd.read_excel('Export_Klamath Falls.xlsx')
```

```python
or_df2
```

```python
df2 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
```

```python
df2['analyte'] = or_df2['Analyte Abbreviation'].str.upper()
df2['concentration'] = or_df2['Final Result'].fillna(0)
df2['sampling_location'] = or_df2['Installation Name']
df2['DL'] = or_df2['Limit of Detection']
df2['sampling_date'] = pd.to_datetime(or_df2['Sample Date'])
df2['sample_id'] = or_df2['Laboratory Sample ID']
df2['units'] = or_df2['Final Units']
df2['state'] = or_df2['State']
df2['analytical_methods'] = or_df2['Analysis Method']
df2['data_source'] = 'DoD'
df2['source_type'] = 'drinking water'
```

```python
df2['is_detected'] = df2['concentration'] != 0
```

```python
df2
```

```python
df = pd.concat([df1, df2])
```

```python
df.index = range(0, len(df))
```
