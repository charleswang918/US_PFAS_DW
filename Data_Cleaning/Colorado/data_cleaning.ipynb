```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
co_d1_nototal
```

```python
df1 = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df1['sample_id'] = co_d1_nototal['SYS_SAMPLE_CODE']
df1['sampling_location'] = co_d1_nototal['LOC_DESC']
df1['data_source'] = 'CDPHE'
df1['sampling_date'] = pd.to_datetime(co_d1_nototal['SAMPLE_DATE'])
df1['pws_id'] = co_d1_nototal['LOC_NAME']
df1['units'] = 'ng/L'
df1['source_type'] = 'drinking water'
df1['state'] = 'Colorado'
df1['MRL'] = co_d1_nototal['REPORT_RESULT_LIMIT']
df1['DL'] = co_d1_nototal['METHOD_DETECTION_LIMIT']
df1['analyte'] = co_d1_nototal['CHEMICAL_NAME'].str.upper()
df1['concentration'] = co_d1_nototal['RESULT_NUMERIC'].fillna(0)
df1['pfas_total'] = fill_list
df1['note'] = co_d1_nototal['REMARK_2']
```

```python
df1['is_detected'] = df1['concentration'] != 0
```
