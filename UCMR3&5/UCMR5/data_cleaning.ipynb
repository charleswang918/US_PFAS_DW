```python
# 分组前的初始化
unique_zipcodes = ucmr5_df['ZIPCODE'].dropna().unique()

# 分成三组
batch_size = len(unique_zipcodes) // 3
batch_1 = unique_zipcodes[:batch_size]
batch_2 = unique_zipcodes[batch_size:2 * batch_size]
batch_3 = unique_zipcodes[2 * batch_size:]
```
