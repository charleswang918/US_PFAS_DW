```python
ppl_df = pd.read_excel('../PFAS Sites and Community Resources.xlsx')
```

```python
ppl_df = ppl_df.dropna(how='all')
```

```python
ppl_df
```

```python
columns = {
    'sample_id': str,
    'sampling_location': str,
    'data_source': str,
    'latitude': float,
    'longitude': float,
    'sampling_date': 'datetime64[ns]',
    'pws_id': str,
    'water_system_name': str,
    'population': int,
    'well_type': str,
    'well_depth': float,
    'units': str,
    'source_type': str,
    'state': str,
    'city': str,
    'county': str,
    'MRL': float,
    'DL': float,
    'analytical_methods': str,
    'is_detected': bool,
    'analyte': str,
    'concentration': float,
    'pfas_total': float,
    'note': str   
}
```

```python
analyte_list = ['PFOA', 'PFOS', 'PFOA+PFOS']
```

```python
columns_to_repeat = ['Site Name', 'State', 'Industry', 'Sample matrix', 'Sample date', 'Total PFAS from one sample (ppt)']
df_repeated = pd.DataFrame(columns=columns_to_repeat)

for i in range(len(ppl_df)):
    row_repeated = pd.concat([ppl_df.iloc[[i]][columns_to_repeat]] * len(analyte_list), ignore_index=True)
    df_repeated = pd.concat([df_repeated, row_repeated], ignore_index=True)

df_repeated
```

```python
df_repeated['analyte'] = analyte_list * len(ppl_df)
```

```python
df_repeated
```

```python
ppl_df = ppl_df.replace('<2', 0)
```

```python
ppl_df = ppl_df.replace('>70', 70)
```

```python
ppl_df[['PFOA (ppt)', 'PFOS (ppt)', 'PFOA+PFOS from one sample (ppt)']] = ppl_df[['PFOA (ppt)', 'PFOS (ppt)', 'PFOA+PFOS from one sample (ppt)']].astype(float)
```

```python
df_repeated['concentration'] = ppl_df[['PFOA (ppt)', 'PFOS (ppt)', 'PFOA+PFOS from one sample (ppt)']].values.flatten().tolist()
```

```python
def set_is_detected(concentration):
    if pd.isna(concentration):
        return np.nan
    return concentration != 0
```

```python
df_repeated['is_detected'] = df_repeated['concentration'].apply(set_is_detected)
```

```python
df_repeated
```

```python
df = pd.DataFrame({col: pd.Series(dtype=dt) for col, dt in columns.items()})
df['analyte'] = df_repeated['analyte'].str.upper()
df['concentration'] = df_repeated['concentration']
df['sampling_location'] = df_repeated['Site Name']
df['sampling_date'] = df_repeated['Sample date'].astype('Int64')
df['units'] = 'ng/L'
df['state'] = df_repeated['State']
df['data_source'] = 'PPL'
df['source_type'] = df_repeated['Sample matrix']
df['is_detected'] = df_repeated['is_detected']
df['pfas_total'] = df_repeated['Total PFAS from one sample (ppt)']
```

```python
df.to_csv('ppl.csv')
```
